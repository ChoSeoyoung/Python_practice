1. 공공 데이터 과거 판매가격(주유소)를 pandas로 읽고 필요한 라이브러리를 불러오시오.

```python
import pandas as pd
df=pd.read_csv('과거 판매가격(주유소).csv')
df.head()
```

2. null 값이 있는지 합계를 확인한다.

```python
df.isnull().sum()
```

3. 누락 데이터가 들어 있는 행을 0으로 채운다.

```python
df=df.fillna(0)
```

4. null 값이 있는지 다시 확인한다.

```python
df.isnull().sum()
```

5. data를 확인한다.

```python
df
```

6. 필요없는 첫번째 행을 삭제하고 확인한다.

```python
df.set_index('번호',inplace=True)
df.drop('기준 : 월간(202111~202111)',axis=0,inplace=True)
df.reset_index(inplace=True)
df
```

7. 데이터프레임의 기술 통계 정보를 요약하여 출력한다.

8. 고급휘발유 값이 0인 데이터 상위 10개만 출력한다.

```python
df_고급휘발유=df[df['고급휘발유']==0]
df_고급휘발유.head(10)
```

9. 고급 휘발유 값이 0인 것만 빼고 다시 저장한다.

```python
df=df[df['고급휘발유']!=0]
```

10. 데이터프레임의 기술 통계 정보를 요약하여 다시 출력한다.

```python
df.describe(include='all')
```

11. count 열을 만들어 상표별로 합계를 구하여 출력한다.

```python
df_상표=df.groupby(['상표']).sum()
```

```python
df_count=df.groupby(['상표']).count()['번호']
```

```python
df_상표['count']=df_count
```

```python
df_상표
```

12. matplotlib 한글 폰트 오류를 해결하도록 처리하고, 다음과 같이 상표별 주유소 개수에 대한 그래프를 그리시오.

```python
import pandas as pd
import matplotlib.pyplot as plt

#matplotlib한글 폰트 오류 문제 해결
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:\\Windows\\Fonts\\malgun.ttf").get_name()
rc('font', family=font_name)

#스타일 서식 지정
plt.style.use('ggplot')

df_상표['count'].plot(kind='pie',
                    autopct='%d%%',
                    figsize=(12,5),
                   startangle=10,
                   colors=['red','green','skyblue','blue','brown','chocolate','bisque','cadetblue'])

plt.title('상표별 주유소 개수')
plt.axis('equal')
plt.legend(labels=df_상표.index,loc='upper right') #범례 표시
plt.show()
```
