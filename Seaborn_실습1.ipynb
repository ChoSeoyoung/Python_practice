## regplot()-회귀선이 있는 산점도
*서로 다른 2개의 연속 변수 사이의 산점도를 그리고 선형회귀분석에 의한 회귀선을 함께 나타낸다.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

남북한발전량=pd.read_excel('./6.시각화도구/남북한발전전력량.xlsx')
남북한발전량
```

남한과 북한의 합계 발전량 사이의 산점도를 그리고 선형회귀분석에 의한 회귀선과 함께 나타낸다.

```python
import matplotlib.pyplot as plt
import seaborn as sns

#matplotlib한글 폰트 오류 문제 해결
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

#데이터셋 가져오기
남북한발전량=pd.read_excel('./6.시각화도구/남북한발전전력량.xlsx')
남북한합계=남북한발전량.iloc[[0,5],2:]
남북한합계.index=['South','North']
납북한합계=남북한합계.T
납북한합계['South']=납북한합계['South'].map(int)
납북한합계['North']=납북한합계['North'].map(int)
납북한합계

#스타일 테마 설정(5가지: darkgrid, whitegrid, dark, white, ticks)
sns.set_style('darkgrid')

#그래프 객체 생성(figure에 2개의 서브 플롯 생성)
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

sns.regplot(x=납북한합계['South'],y=납북한합계['North'],ax=ax1)
ax2=sns.regplot(x=납북한합계['South'],y=납북한합계['North'],fit_reg=False,color='orange')

plt.show()
```

남한읫수력 화력 발전량 사이의 산점도를 그리고 선형회귀분석에 의한 회귀선과 함께 나타낸다.

```python
import matplotlib.pyplot as plt
import seaborn as sns

#matplotlib한글 폰트 오류 문제 해결
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

#데이터셋 가져오기
남북한발전량=pd.read_excel('./6.시각화도구/남북한발전전력량.xlsx')
남한수력화력=남북한발전량.iloc[[1,2],2:]
남한수력화력.index=['Water','Fire']
남한수력화력=남한수력화력.T
남한수력화력['Water']=남한수력화력['Water'].map(int)
남한수력화력['Fire']=남한수력화력['Fire'].map(int)

#스타일 테마 설정(5가지: darkgrid, whitegrid, dark, white, ticks)
sns.set_style('whitegrid')

#그래프 객체 생성(figure에 2개의 서브 플롯 생성)
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

sns.regplot(x=남한수력화력['Water'],y=남한수력화력['Fire'],ax=ax1)
ax2=sns.regplot(x=남한수력화력['Water'],y=남한수력화력['Fire'],fit_reg=False,color='orange')

plt.show()
```

남한의 수력발전량이 증가할수록 화력발전량이 증가하는 경향을 확인할 수 있다.

## jointplot()-산점도를 기본으로 표시하고, x-y축에 대한 히스토그램을 동시에 보여준다.

```python
import matplotlib.pyplot as plt
import seaborn as sns

#matplotlib한글 폰트 오류 문제 해결
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

#데이터셋 가져오기
남북한발전량=pd.read_excel('./6.시각화도구/남북한발전전력량.xlsx')
남한수력화력=남북한발전량.iloc[[1,2],2:]
남한수력화력.index=['Water','Fire']
남한수력화력=남한수력화력.T
남한수력화력['Water']=남한수력화력['Water'].map(int)
남한수력화력['Fire']=남한수력화력['Fire'].map(int)

#스타일 테마 설정(5가지: darkgrid, whitegrid, dark, white, ticks)
sns.set_style('white')

sns.jointplot(x='Water',y='Fire',data=남한수력화력)
sns.jointplot(x='Water',y='Fire',kind='reg',data=남한수력화력)
sns.jointplot(x='Water',y='Fire',kind='hex',data=남한수력화력)
sns.jointplot(x='Water',y='Fire',kind='kde',data=남한수력화력)

plt.show()
```

데이터의 선형관계를 파악할 수 있다.

# 타이타닉

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#Seaborn 제공 데이터셋 가져오기
titanic=sns.load_dataset('titanic')

print(titanic.head())
print('\n')
print(titanic.info())
```

## distplot-히스토그램, 커널 밀도그래프

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

sns.distplot(titanic['fare'],ax=ax1)
sns.distplot(titanic['fare'],hist=False,ax=ax2)
sns.distplot(titanic['fare'],kde=False,ax=ax3)
```

0~100사이의 요금에 가장 몰려 있다

## heatmap()-히트맵

```python
table=titanic.pivot_table(index=['sex'],columns=['class'],aggfunc='size')

#히트맵 그리기
sns.heatmap(table, annot=True, fmt='d', cmap='YlGnBu',linewidth=.5,cbar=False)

plt.show()
```

여자(female) 승객보다 남자(male) 승객이 상대적으로 많은 편이고 Third Class 남자 승객의 수가 압도적으로 많은 것을 알 수 있다.

```python
table=titanic.pivot_table(index=['class'],columns=['embarked'],aggfunc='size')

#히트맵 그리기
sns.heatmap(table, annot=True, fmt='d', cmap='YlGnBu',linewidth=.5,cbar=False)

plt.show()
```

3등석, S선착장에서 탑승한 사람이 압도적으로 많다

## stripplot, swarmplot - 범주형 데이터의 산점도

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

#x축,y축에 변수 할당
sns.barplot(x='sex',y='survived',data=titanic,ax=ax1)

#x축,y축에 변수 할당하고 hue옵션 추가하여 누적 출력
sns.barplot(x='sex',y='survived',hue='class',data=titanic,ax=ax2)

#x축,y축 변수 할당하고 hue옵션을 추가하여 누적 출력
sns.barplot(x='sex',y='survived',hue='class',dodge=False,data=titanic,ax=ax3)

#차트 제목 표시
ax1.set_title('titanic survived - sex')
ax2.set_title('titanic survived - sex/class')
ax3.set_title('titanic survived - sex/class(stacked)')

plt.show()
```

여성의 생존률이 높고, class를 나눠서 비교해보면 1등석일수록 생존률이 높다.

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

#x축,y축에 변수 할당
sns.barplot(x='class',y='survived',data=titanic,ax=ax1)

#x축,y축에 변수 할당하고 hue옵션 추가하여 누적 출력
sns.barplot(x='class',y='survived',hue='embarked',data=titanic,ax=ax2)

#x축,y축 변수 할당하고 hue옵션을 추가하여 누적 출력
sns.barplot(x='class',y='survived',hue='embarked',dodge=False,data=titanic,ax=ax3)

#차트 제목 표시
ax1.set_title('titanic survived - class')
ax2.set_title('titanic survived - class/embarked')
ax3.set_title('titanic survived - class/embarked(stacked)')

plt.show()
```

## countplot()-빈도그래프

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

#x축,y축에 변수 할당
sns.countplot(x='class',palette='Set1',data=titanic,ax=ax1)

#x축,y축에 변수 할당하고 hue옵션 추가하여 누적 출력
sns.countplot(x='class',palette='Set2',hue='sex',data=titanic,ax=ax2)
#x축,y축 변수 할당하고 hue옵션을 추가하여 누적 출력
sns.countplot(x='class',palette='Set3',hue='sex',dodge=False,data=titanic,ax=ax3)
#차트 제목 표시
ax1.set_title('titanic class')
ax2.set_title('titanic class - who')
ax3.set_title('titanic class - who(stacked)')

plt.show()
```

## boxplot()-박스플롯, violinplot()-바이올린플롯

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

#x축,y축에 변수 할당
sns.boxplot(x='survived',y='age',data=titanic,ax=ax1)

#x축,y축에 변수 할당하고 hue옵션 추가하여 누적 출력
sns.boxplot(x='survived',y='age',hue='sex',data=titanic,ax=ax2)

plt.show()
```

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

#x축,y축에 변수 할당
sns.violinplot(x='survived',y='age',data=titanic,ax=ax1)

#x축,y축에 변수 할당하고 hue옵션 추가하여 누적 출력
sns.violinplot(x='survived',y='age',hue='sex',data=titanic,ax=ax2)

plt.show()
```
