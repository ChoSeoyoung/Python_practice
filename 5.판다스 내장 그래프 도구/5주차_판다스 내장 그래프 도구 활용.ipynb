# 판다스 내장 그래프 도구
판다스 그래프 도구 사용 방법
-시리즈 또는 데이터프레임 객체에 plot() 메소드를 적용하고, kind옵션으로 그래프 종류를 선택한다.
* line: 선그래프
* bar: 수직막대그래프
* barh: 수평막대그래프
* his: 히스토그램
* box: 박스플롯
* kde: 커널 밀도 그래프
* area: 면적 그래프
* pie: 파이 그래프
* scatter: 산점도 그래프
* hexbin: 고밀도 산점도 그래프

DataFrame.plot(kind="line",x=column,y=columns,color=color)

### 예제 3-4. 선그래프

```python
import pandas as pd

df = pd.read_excel('남북한발전전력량.xlsx')

#남한, 북한 발전량 합계 데이터만 추출
df_ns = df.iloc[[0,5],2:]
df_ns.index=["South","North"]
#열 이름의 자료형을 정수형으로 변경
df_ns.columns=df_ns.columns.map(int)
print(df_ns)
print('\n')

# 선 그래프 그리기
df_ns.plot()
```

```python
#행과 열을 전치하여 다시 그리기
df_ns = df_ns.T
print(df_ns)
print('\n')

df_ns.plot()
```

### 예제 3-5. 막대그래프

```python
import pandas as pd

df = pd.read_excel('남북한발전전력량.xlsx')

#남한, 북한 발전량 합계 데이터만 추출
df_ns = df.iloc[[0,5],2:]
df_ns.index=["South","North"]
#열 이름의 자료형을 정수형으로 변경
df_ns.columns=df_ns.columns.map(int)

#행, 열 전치하여 막대 그래프 그리기
df_ns=df_ns.T
print('df_ns')
print('\n')

df_ns.plot(kind="bar")
```

```python
#누적막대그래프
df_ns.plot(kind="bar", color=["green","blue"], stacked=True)
```

```python
#가로 막대 그래프
df_ns.plot(kind="barh")
```

```python
#가로 누적 막대 그래프
df_ns.plot(kind="barh",stacked=True)
```

### 예제 3-6. 히스토그램

```python
import pandas as pd

df = pd.read_excel('남북한발전전력량.xlsx')

#남한, 북한 발전량 합계 데이터만 추출
df_ns = df.iloc[[0,5],2:]
df_ns.index=["South","North"]
#열 이름의 자료형을 정수형으로 변경
df_ns.columns=df_ns.columns.map(int)

#행, 열 전치하여 막대 그래프 그리기
df_ns=df_ns.T
print('df_ns')
print('\n')

df_ns.plot(kind="hist")
```

### 예제 3-7. 산점도

```python
import pandas as pd

df = pd.read_csv('auto-mpg.csv')
df.plot(kind='scatter',x='weight',y='mpg')
```

```python
# 2개의 열을 선택하여 산점도 그리기
# 점의 크기와 색 설정
df.plot(kind='scatter',x='weight',y='mpg', color='red', s=5)
```

### 예제3-8. 박스플롯

```python
import pandas as pd

df = pd.read_csv('auto-mpg.csv')
df[['mpg','cylinders']].plot(kind="box")
```
