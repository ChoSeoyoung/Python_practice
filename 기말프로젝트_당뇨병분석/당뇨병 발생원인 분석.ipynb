# 지방지표

```python
지방지표=pd.read_csv('e지방지표.csv',encoding='cp949')
지방지표
```

```python
def func_지방지표(temp,name):
    #2015년
    temp_2015=pd.DataFrame({name:temp['2015'],
                '시도':list(temp['시도'].unique()),
                  '연도':[2015]*len(list(temp['시도'].unique()))})
    temp_2015 = temp_2015.set_index(['연도','시도'])

    #2016년
    temp_2016=pd.DataFrame({name:temp['2016'],
                '시도':list(temp['시도'].unique()),
                  '연도':[2016]*len(list(temp['시도'].unique()))})
    temp_2016 = temp_2016.set_index(['연도','시도'])

    #2017년
    temp_2017=pd.DataFrame({name:temp['2017'],
                '시도':list(temp['시도'].unique()),
                  '연도':[2017]*len(list(temp['시도'].unique()))})
    temp_2017 = temp_2017.set_index(['연도','시도'])

    #2018년
    temp_2018=pd.DataFrame({name:temp['2018'],
                '시도':list(temp['시도'].unique()),
                  '연도':[2018]*len(list(temp['시도'].unique()))})
    temp_2018 = temp_2018.set_index(['연도','시도'])

    #2019년
    temp_2019=pd.DataFrame({name:temp['2019'],
                '시도':list(temp['시도'].unique()),
                  '연도':[2019]*len(list(temp['시도'].unique()))})
    temp_2019 = temp_2019.set_index(['연도','시도'])

    result=pd.concat([temp_2015,temp_2016,temp_2017,temp_2018,temp_2019])
    return result
```

## 지방지표_개인소득

```python
지방지표_개인소득=지방지표[지방지표['지표별(2)']=='1인당 개인소득 (천원)']
지방지표_개인소득.drop(columns=['지표별(1)','지표별(2)'],inplace=True)
지방지표_개인소득.columns=['시도','2015','2016','2017','2018','2019']
개인소득=func_지방지표(지방지표_개인소득,'개인소득')
개인소득['개인소득']=pd.to_numeric(개인소득['개인소득'])
개인소득['개인소득']=개인소득['개인소득']/개인소득['개인소득'].max()
개인소득
```

```python
개인소득['개인소득']=개인소득['개인소득'].astype(float)
개인소득_지역=개인소득.groupby(['시도'])['개인소득'].mean()
개인소득_지역=pd.DataFrame(개인소득_지역)
개인소득_지역.reset_index(inplace=True)
개인소득_지역.loc[len(개인소득_지역)]=['전국',개인소득_지역['개인소득'].mean()]
개인소득_지역=개인소득_지역.sort_values(by=['개인소득'],axis=0)

개인소득_지역_전국=개인소득_지역[개인소득_지역['시도']=='전국']
개인소득_지역_평균이상=개인소득_지역[개인소득_지역['개인소득']>float(개인소득_지역[개인소득_지역['시도']=='전국']['개인소득'])]

#그래프그리기
plt.bar('시도','개인소득', data=개인소득_지역 ,color='peachpuff')
plt.bar('시도','개인소득', data=개인소득_지역_전국 ,color='lightgray')
plt.bar('시도','개인소득', data=개인소득_지역_평균이상 ,color='salmon')
plt.xticks(rotation='vertical')
plt.title("개인소득_지역별")
```
