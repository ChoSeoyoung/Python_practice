# 시도별 전출입 인구수

```python
df = pd.read_excel('./6.시각화도구/시도별 전출입 인구수.xlsx',header=0)
df.set_index('전출지별',inplace=True)
df.drop('전출지별',axis=0,inplace=True)
df.reset_index(inplace=True)
df.fillna(method='ffill',inplace=True)
df
```

각 시도별 전입인구 수 비교

```python
df_전입=df[df['전출지별']=='전국']
col_years=list(map(str,range(1970,2018)))
df_전입.drop('전출지별',axis=1,inplace=True)
df_전입.rename({'전입지별':'전입지'},axis=1,inplace=True)
df_전입.set_index('전입지',inplace=True)
df_전입=df_전입.T
df_전입.plot(figsize=(10,6))

#제목 지정
plt.title('시도별 전입인구수')

#x축, y축 라벨 지정
plt.xlabel('연도')
plt.ylabel('전입인구수')

plt.legend(df_전입.columns,loc='upper right')

plt.show()
```

특히 어느 도시에서 서울로 많이 전입하는지 알아보자.

```python
df = pd.read_excel('./6.시각화도구/시도별 전출입 인구수.xlsx',header=0)
df.set_index('전출지별',inplace=True)
df.drop('전출지별',axis=0,inplace=True)
df.reset_index(inplace=True)
df=df.fillna(method='ffill')

df_서울전입=df[(df['전입지별']=='서울특별시')&(df['전출지별']!='서울특별시')]
df_서울전입.drop('전입지별',axis=1,inplace=True)
df_서울전입.rename(columns={'전출지별':'전출지'},inplace=True)
df_서울전입=df_서울전입[df_서울전입['전출지']!='전국']
df_서울전입.set_index('전출지',inplace=True)
df_서울전입=df_서울전입.T
df_서울전입
```

```python
df_서울전입.plot(figsize=(10,6))

plt.title('서울로 전입하는 인구')

plt.legend(df_서울전입.columns,loc='upper right')

plt.xlabel('연도')
plt.ylabel('전입인구수')

plt.show()
```

## 서울->경기 인구 이동

```python
df=df.fillna(method='ffill')
mask = (df['전출지별']=='서울특별시')&(df['전입지별']!='서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop('전출지별',axis=1)
df_seoul.rename(columns={'전입지별':'전입지'},inplace=True)
df_seoul.set_index('전입지',inplace=True)
df_seoul

#서울에서 경기도로 이동한 인구 데이터 값만 선택
sr_one = df_seoul.loc['경기도']

fig=plt.figure(figsize=(15,5))

#x축, y축 데이터를 plot함수에 입력
plt.plot(sr_one,color='orange',marker='o',linewidth=10,markerfacecolor='yellow',markersize=12)

plt.title('서울->경기 인구 이동')

#축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')

#x축 눈금 라벨 회전하기
plt.xticks(rotation=90)

plt.ylim(0,750000)

#범례
plt.legend(labels=['서울->경기'],loc='best')

plt.show() #변경사항 저장하고 그래프 출력
```

## 서울->부산,충남,경상북도

```python
df = pd.read_excel('./6.시각화도구/시도별 전출입 인구수.xlsx',header=0)
df=df.fillna(method='ffill')
mask = (df['전출지별']=='서울특별시')&(df['전입지별']!='서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop('전출지별',axis=1)
df_seoul.rename(columns={'전입지별':'전입지'},inplace=True)
df_seoul.set_index('전입지',inplace=True)
df_seoul

#서울에서 경기도로 이동한 인구 데이터 값만 선택
df_3 = df_seoul.loc[['부산광역시','대구광역시','인천광역시']]
df_3=df_3.replace('-', None)
col_years=list(map(str,range(1970,2018)))

plt.style.use('ggplot')

fig=plt.figure(figsize=(20,5))
ax=fig.add_subplot(1,1,1)

#x축, y축 데이터를 plot함수에 입력
ax.plot(col_years,df_3.loc['부산광역시',:],color='pink',marker='o',linewidth=2,markersize=12,label='서울->부산')
ax.plot(col_years,df_3.loc['대구광역시',:],color='orange',marker='o',linewidth=2,markersize=12,label='서울->대구')
ax.plot(col_years,df_3.loc['인천광역시',:],color='green',marker='o',linewidth=2,markersize=12,label='서울->인천')

ax.set_title('서울->부산,대구,인천 인구 이동')

#축 이름 추가
ax.set_xlabel('기간')
ax.set_ylabel('이동 인구수')

#x축 눈금 라벨 회전하기
ax.set_xticklabels(col_years,rotation=90)

ax.legend(loc='best')

plt.show() #변경사항 저장하고 그래프 출력
```

## 보조축 활용하기

```python
import pandas as pd
import matplotlib.pyplot as plt

#matplotlib한글 폰트 오류 문제 해결
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)


plt.style.use('ggplot') #스타일 서식 지정
plt.rcParams['axes.unicode_minus']=False

#Excel 데이터를 데이터프레임으로 변환
df = pd.read_excel('./6.시각화도구/남북한발전전력량.xlsx',convert_float=True)
df = df.loc[5:9] #남한 데이터만 가져옴
df.drop('전력량 (억㎾h)',axis='columns',inplace=True)
df.set_index('발전 전력별',inplace=True)
df=df.T

#증감률(변동률) 계산
df = df.rename(columns={'합계':'총발전량'})
df['총발전량-1년']=df['총발전량'].shift(1)
df['증감률']=((df['총발전량']/df['총발전량-1년'])-1)*100

#2축 그래프 그리기
ax1=df[['수력','화력']].plot(kind='bar',figsize=(20,10),width=0.7,stacked=True,color=['cornflowerblue','orange'])
ax2=ax1.twinx()
ax2.plot(df.index,df.증감률,ls='--',marker='o',markersize=20,color='green',label='전년대비 증감률(%)')

ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)

ax1.set_xlabel('연도',size=20)
ax1.set_ylabel('발전량(억 KWh)')
ax2.set_ylabel('전년 대비 증감률(%)')

plt.title('북한 전력 발전량 (1990~2016)',size=30)
ax1.legend(loc='upper left')

plt.show()
```
